```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def fn_mx(x):
    if x >= -1 and x <= 0:
        return x + 1
    elif x > 0 and x <= 1:
        return -x + 1
```

```python
def fn_my(y):
    if y >= -1 and y <= 0:
        return y + 1
    elif y > 0 and y <= 1:
        return 1 - y
```

```python
l = np.linspace(-1, 1, 100)
marginal_l1 = list(map(lambda x: fn_mx(x), l))
marginal_l2 = list(map(lambda y: fn_my(y), l))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(l, marginal_l1, '-b')
axes[0].set_xlabel(r"$\lambda_1$")
axes[0].set_ylabel(r"$f_{x}(\lambda_1)$")
axes[0].set_title(r'Marginal density for $\lambda_1$')

axes[1].plot(l, marginal_l2, '-b')
axes[1].set_xlabel(r"$\lambda_2$")
axes[1].set_ylabel(r"$f_{x}(\lambda_2)$")
axes[1].set_title(r'Marginal density for $\lambda_2$')
plt.show()
```
